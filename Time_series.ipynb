Szereg czasowy dla danych hotele: Ilość gości hotelowych na dolym slasku w okresie 2009 a 2014

https://boostedml.com/2020/05/visualizing-time-series-in-r.html link z przykładami


```r
install.packages('forecast', dependencies = TRUE)
install.packages('ggfortify')
install.packages('tseries')
install.packages('ggplot2')
install.packages('astsa')

library(ggfortify)
library(tseries)
library(astsa)
library(forecast)
library(ggplot2)
```

```r
hotele <- read.table("hotele.txt")
head(hotele)
```

```r
hotele.t = ts(hotele,start= c(2009,1),frequency = 12)
hotele.t
```

```r
plot(hotele.t)
```

```r
ggsubseriesplot(hotele.t)
```

```r
seasonplot(hotele.t, year.labels = T, col = rainbow(6))
```

```r
acf(hotele.t)
```

```r
plot(decompose(hotele.t))
```

Wnioski: widzimy sezonowośc pojawiajaca sie w okresie wakacji oraz zimowym

Widoczny jest trend wzrostowy
